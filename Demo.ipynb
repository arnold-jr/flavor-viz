```python
%matplotlib
import pandas as pd
```

```python
cf = pd.read_csv('compounds_flavors.tsv', delimiter='|')
```

```python
cf.info()
```

```python
cf.head()
```

```python
cf.compound_id.unique().size
```

```python
cf.groupby('compound_id').aggregate('count')
```

```python
cf_pivot = cf\
  .pivot(index='compound_id', columns='flavor_name', values='flavor_id')\
  .rename_axis(None)\
  .fillna(0)\
  .applymap(lambda x: 0.0 if x == 0.0 else 1.0)
cf_pivot
```

```python
from sklearn.manifold import TSNE

model = TSNE(n_components=1, random_state=0)
tsne = model.fit_transform(cf_pivot.as_matrix().transpose())
```

```python
tsne_ind = tsne.ravel().argsort()
ordered_cols = cf_pivot.columns.to_series().iloc[tsne_ind].tolist()
cf_final = cf_pivot[ordered_cols]
```

```python
cf_final
```

```python
fc = pd.read_csv('foods_compounds.tsv', delimiter='|').drop_duplicates(subset=['compound_id', 'food_common_name'])
fc.info()
```

```python
fc.head()
```

```python
fc.groupby('unit').aggregate('count')
```

```python
fc.unit.unique()
```

```python
unit_conversion = {
  'g': 1e3, 
  'mg': 1.0, 
  'NE': 1.0, 
  'ug': 1e-3, 
  'RE': 1.0, 
  'aTE': 1.0, 
  'IU': 1.0, 
  'ug_RAE': 1.0, 
  'mg/100 g': 1.0, 
  'mg/100 ml': 1.0, 
  'mg/100g': 1.0, 
  'Î±-TE': 1.0, 
  'Âµg': 1e-3, 
  'ug/g': 1e-3
}
```

```python
fc['content'] = fc.content.multiply(fc.unit.apply(lambda k: unit_conversion[k]))
```

```python
fc_pivot = fc.pivot(index='food_common_name', columns='compound_id', values='content').rename_axis(None).fillna(0.0)
```

```python
fc_pivot.info()
```

```python
common = fc_pivot.columns.intersection(cf_final.index).tolist()
common
```

```python
left = fc_pivot[common]
right = cf_final.loc[common, :]
ff = left.dot(right)
ff
```

```python
ff.info()
```

```python
ff = ff.iloc[:, list(reversed(ff.mean().argsort()))]
```

```python
ff.sample(10)
```

```python
import matplotlib.pyplot as plt

#ax = plt.subplot(111, frame_on=False) # no visible frame
#ax.xaxis.set_visible(False)  # hide the x axis
#ax.yaxis.set_visible(False)  # hide the y axis

img = ff.as_matrix()
img = img.clip(max=img.ravel().mean()+3.0*img.ravel().std())
plt.imshow(img, cmap='gray')
```
